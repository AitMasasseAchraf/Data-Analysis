```python
import pandas as pd
import re
import string
import nltk
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from textblob import TextBlob
import numpy
nltk.download('stopwords')
from nltk.corpus import stopwords
ps = nltk.PorterStemmer()
wn = nltk.WordNetLemmatizer()
stop_words = stopwords.words('english')

def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

def read_file(path):
    data1=pd.read_csv(path)
    return data1

def tokenization(text):
    text = re.split('\W+', text)
    return text

def remove_stopwords(text):
    text = [word for word in text if word not in stop_words]
    return text

def stemming(text):
    text = [ps.stem(word) for word in text]
    return text

def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return text

def analyser_sentiment(Data_Frame):
    noOfTweet=3951
    positive = 0
    negative = 0
    neutral = 0
    polarity = 0
    tweet_list = []
    neutral_list = []
    negative_list = []
    positive_list = []
    lis=[]
    posi=0
    nega=0
    neut=0
    for i in range(0,len(Data_Frame.index)) :
        for tweet in Data_Frame.text.iloc[i]:
            tweet_list.append(tweet)
            analysis = TextBlob(tweet)
            score = SentimentIntensityAnalyzer().polarity_scores(tweet)
            if score['compound']<0:
               posi+=1
            elif score['compound']>0:
                nega+=1
              
            elif score['compound']==0:
                neut+=1
            
         
        if posi>nega and posi>neut:
                 lis.append("positive")
        elif nega>posi and nega>neut:
                  lis.append("negative")
        elif neut>posi and neut>nega:
            lis.append("neutral")
            

    return lis

def export_dataframe(path_puredata,path_result,Data_Frame):
    data1=read_file(path_puredata)
    data1["type_tweet"]=analyser_sentiment(Data_Frame)
    data1.to_csv(path_result, index = False)
  

path1='C:\\Users\\aitma\\Downloads\\data_1.txt'
path2='C:\\Users\\aitma\\Desktop\\my_scripts\\data2_ex.csv'
data1=read_file(path1)
    
data1['text'] = data1['text'].apply(lambda x: remove_punct(x))

data1['text'] = data1['text'].apply(lambda x: tokenization(x.lower()))
data1['text'] = data1['text'].apply(lambda x: remove_stopwords(x))
data1['text'] = data1['text'].apply(lambda x: stemming(x))
data1['text'] = data1['text'].apply(lambda x: lemmatizer(x))
export_dataframe(path1,path2,data1)
```
